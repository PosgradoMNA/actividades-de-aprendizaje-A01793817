# **6. Escala los datos, si consideras necesario**

```python
from sklearn.preprocessing import StandardScaler

# Cargar datos de las características
X = df_new.drop('Y', axis='columns')

X = StandardScaler().fit_transform(X)
X = pd.DataFrame(X)
X.head()
```

# **7. Reduce las dimensiones con PCA, si consideras necesario.**
* Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
* Indica la importancia de las variables en cada componente

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 24)
x_pca = pca.fit_transform(X) 
x_pca = pd.DataFrame(x_pca) 
x_pca.head()

#print(pca.explained_variance_ratio_)
#print(("Explained Variance: %s") % (fit.explained_variance_ratio_))
#print(fit.components_)
```

```python
varianza_explicada = pca.explained_variance_
varianza_explicada
```

```python
import matplotlib.pyplot as plt

x_pca.hist(color="blue", ec="black", figsize=(25,15))
plt.title("Distribución de los atributos")
plt.show()
```

# **9. Realiza la visualización de los datos usando por lo menos 3 gráficos que consideres adecuados: plot, scatter, jointplot, boxplot, areaplot, pie chart, pairplot, bar chart, etc.**

```python
import pandas as pd
import matplotlib.pyplot as plt

df_bar=df_new
df_bar['X3']=df['X3'].replace([1,2,3,4,5,6,0],['posgrado','universidad','secundaria','otros','otros','otros','otros'])
serie_est=df_bar.X3.value_counts()
serie_est.index
serie_est.values


plt.bar(x=serie_est.index, height=serie_est.values)
plt.xlabel('Nivel Educativo', fontsize=10)
plt.ylabel('Crédito', fontsize=10)
plt.title("Monto de crédito otorgado según nivel educativo", fontsize=15)
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_pie=df_new

df_pie["X2"] = df_pie["X2"].astype("int32")
df_group=df_pie.groupby(["X2"])["X5"].count().reset_index()
df_group.columns=["X2","number"]

plt.pie(df_group["number"],labels=df_group["X2"].replace([1,2],["Masculino","Femenino"]),autopct="%.2f")

plt.title("% en género", fontsize=15)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_sca=df_new


plt.scatter(x=df_sca.X5, y=df_sca.X1, color="red")
plt.xlabel('Edad', fontsize=10)
plt.ylabel('Crédito', fontsize=10)
plt.title("Crédito según la edad", fontsize=15)
plt.show()
```
